Exploring team-wise performance metrics involves analyzing various aspects of teams' performances in a cricket tournament. Here's a breakdown of steps using Python and Pandas to explore these metrics:

```python

# Sum of runs and wickets per team
team_metrics_sum = data.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum'
}).reset_index()
team_metrics_sum.columns = ['team', 'total_score', 'total_wickets']

# Mean economy rate per team
team_metrics_mean_econ = data.groupby('team').agg({
    'econ': 'mean'
}).reset_index()
team_metrics_mean_econ.columns = ['team', 'avg_economy']

# Max and Min runs scored by a team
team_max_runs = data.groupby('team')['runs'].max().reset_index()
team_max_runs.columns = ['team', 'max_run']

# Batting and Bowling styles - Calculate averages
team_batting_style = team_metrics_sum['total_score'] / team_metrics_sum['total_wickets']
team_bowling_style = team_metrics_sum['total_wickets'] / team_metrics_mean_econ['avg_economy']

# Create a dataframe with calculated batting and bowling styles
team_styles = pd.DataFrame({
    'team': team_metrics_sum['team'],
    'batting_style': team_batting_style,
    'bowling_style': team_bowling_style
})

# Combine various metrics into a summary dataframe
team_summary = pd.merge(team_metrics_sum, team_metrics_mean_econ, on='team')
team_summary = pd.merge(team_summary, team_styles, on='team')
team_summary = pd.merge(team_summary, team_max_runs, on='team')

# Display the combined summary of team-wise metrics with batting and bowling styles
print("\nCombined Team-wise Metrics Summary:")
print(team_summary)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'team' column is used as the team identifier
teams = team_summary['team']
total_runs = team_summary['total_score']
total_wickets = team_summary['total_wickets']
avg_economy = team_summary['avg_economy']
batting_style = team_summary['batting_style']
bowling_style = team_summary['bowling_style']
# max_runs = team_summary['max_run']

# Normalize the data for better visualization
# total_runs_normalized = total_runs / max_runs.max()
total_wickets_normalized = total_wickets / total_wickets.max()
avg_economy_normalized = avg_economy / avg_economy.max()
batting_style_normalized = batting_style / batting_style.max()
bowling_style_normalized = bowling_style / bowling_style.max()

# Set up positions for bars
bar_width = 0.15
bar_positions = np.arange(len(teams))

# Plotting grouped bar chart
plt.figure(figsize=(14, 8))
# plt.bar(bar_positions - 2*bar_width, total_runs_normalized, width=bar_width, label='Total Runs', alpha=0.8)
plt.bar(bar_positions - bar_width, total_wickets_normalized, width=bar_width, label='Total Wickets', alpha=0.8)
plt.bar(bar_positions, avg_economy_normalized, width=bar_width, label='Avg Economy', alpha=0.8)
plt.bar(bar_positions + bar_width, batting_style_normalized, width=bar_width, label='Batting Style', alpha=0.8)
plt.bar(bar_positions + 2*bar_width, bowling_style_normalized, width=bar_width, label='Bowling Style', alpha=0.8)

# Customize plot
plt.title('Combined Team-wise Metrics Summary')
plt.xlabel('Team')
plt.ylabel('Normalized Values')
plt.xticks(bar_positions, teams, rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```

## Batting Statistics per Player

```python
batting_stats = data[data['bat_or_bowl'] == 'bat'].groupby('player').agg({
    'runs': 'sum',
    '4s': 'sum',
    '6s': 'sum'
}).reset_index()

print("Batting Statistics per Player:")
print(batting_stats)
```

## Bowling Statistics per Player

```python
bowling_stats = data[data['bat_or_bowl'] == 'bowl'].groupby('player').agg({
    'wkts': 'sum',
    'econ': 'mean'
}).reset_index()

print("\nBowling Statistics per Player:")
print(bowling_stats)
```

```python
import matplotlib.pyplot as plt

# Assuming 'player' column is used as the player identifier
players = bowling_stats['player']
wkts = bowling_stats['wkts']
econ = bowling_stats['econ']

# Increase figure size and adjust font size
plt.figure(figsize=(12, 20))

# Plotting horizontal bar chart for wickets
plt.barh(players, wkts, color='blue')
plt.title('Total Wickets per Player', fontsize=16)
plt.xlabel('Total Wickets', fontsize=14)
plt.ylabel('Player', fontsize=14)
plt.xticks(fontsize=9)
plt.yticks(fontsize=9)
plt.tight_layout()
plt.show()

# Plotting horizontal bar chart for average economy rate
plt.figure(figsize=(12, 20))

plt.barh(players, econ, color='green')
plt.title('Average Economy Rate per Player', fontsize=16)
plt.xlabel('Average Economy Rate', fontsize=14)
plt.ylabel('Player', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()
```
